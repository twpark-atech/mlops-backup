```python
import boto3

s3 = boto3.client(
    "s3",
    endpoint_url="http://localhost:9000",   # MinIO 주소 (WSL에서 docker 쓰면 이거)
    aws_access_key_id="minio",
    aws_secret_access_key="miniostorage",
    region_name="us-east-1",  # 아무거나
)

# 버킷 목록 보기
resp = s3.list_buckets()
for b in resp["Buckets"]:
    print(b["Name"])
```

```python
resp = s3.list_objects_v2(Bucket="its", Prefix="traffic/silver/")
for obj in resp.get("Contents", []):
    print(obj["Key"])
```

```python
import pandas as pd
import pyarrow.fs as pafs

fs = pafs.S3FileSystem(
    access_key="minio",
    secret_key="miniostorage",
    endpoint_override="http://localhost:9000",
    scheme="http",   # MinIO http라면
)

path = "its/traffic/silver/date=20251113/part-00000-f82e3cce-e40a-44ae-808e-189160969be1-c000.snappy.parquet"

# ✨ 포인트: fs.open()으로 파일 핸들 열어서 읽기
with fs.open_input_file(path) as f:
    df = pd.read_parquet(f)  # 여기엔 filesystem 안 줌 (그냥 파일 객체)

print(df.head())
print(df.dtypes)
```
